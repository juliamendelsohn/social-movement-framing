```python
import pandas as pd
import os
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
train_set = '/home/juliame/social-movements/annotated_data/data_splits_06-21-2023/train_full.tsv'
test_set = '/home/juliame/social-movements/annotated_data/data_splits_06-21-2023/test_full.tsv'
df_train = pd.read_csv(train_set,sep='\t')
df_test = pd.read_csv(test_set,sep='\t')
df = pd.concat([df_train,df_test])
```

```python
df['Task2:stance'] = df['Task2:stance'].replace({0:'conservative',1:'neutral',2:'progressive'})
df = df[df['Task1:relevance']==1]
df = df.drop(columns=['id_str','Task1:relevance','Task2:stance-explicit'])
```
